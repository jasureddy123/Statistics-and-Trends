```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the drive module from Google Colab
from google.colab import drive
# Mount Google Drive to access files stored in your Google Drive
drive.mount('/content/drive')
```

```python
world_population = pd.read_csv("/content/drive/MyDrive/World_GDP_Population_CO2_Emissions_Dataset.csv")
```

```python
# Display summary information about the DataFrame
world_population.info()
```

```python
# Display summary statistics of numerical columns in the DataFrame
print(world_population.describe())
```

```python
# Remove rows with missing values
world_population.dropna(inplace=True)
# Remove duplicate rows
world_population.drop_duplicates(inplace=True)
```

# Statistical Moments


```python
# Statistical Moments
mean_values = world_population.mean()
print("Mean:\n", mean_values)
```

```python
median_values=world_population.median()
print("Median:\n", median_values)
```

```python
variance_values = world_population.var()
print("Variance:\n", variance_values)
```

```python
skewness_values = world_population.skew()
print("Skewness:\n", skewness_values)
```

```python
kurtosis_values = world_population.kurt()
print("Kurtosis:\n", kurtosis_values)
```

```python
skewness_values = world_population.skew()
print("Skewness:\n", skewness_values)
```

# Relational Plots


```python
#line Graph
plt.figure(figsize=(10, 5))
sns.lineplot(x=world_population['Year'], y=world_population['GDP Real (USD) '])
plt.title("GDP Over the Years")
plt.xlabel("Year")
plt.ylabel("GDP (USD)")
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(8, 5))
sns.violinplot(y=world_population['GDP growth (%)'])
plt.title("Violin Plot of GDP Growth (%)")
plt.ylabel("GDP Growth (%)")
plt.grid(True)
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(world_population.corr(), annot=True, cmap="coolwarm", fmt='.2f')
plt.title("Feature Correlation Matrix")
plt.show()
```
